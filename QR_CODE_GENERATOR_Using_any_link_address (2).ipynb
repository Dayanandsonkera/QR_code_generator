```python
import qrcode

# Input data for the QR code
data = input("Enter the text or URL to generate a QR code: ")

# Create QR Code object
qr = qrcode.QRCode(
    version=1,  # Controls the size of the QR Code (1 is 21x21, 40 is 177x177)
    error_correction=qrcode.constants.ERROR_CORRECT_L,  # Error correction level
    box_size=10,  # Size of each box in pixels
    border=4,  # Thickness of the border (minimum is 4)
)

# Add data to the QR Code
qr.add_data(data)
qr.make(fit=True)

# Generate QR Code image
qr_image = qr.make_image(fill_color="black", back_color="white")

# Save the image to a file
filename = "qr_code.png"
qr_image.save(filename)

print(f"QR Code generated and saved as '{filename}'.")
```

```python
import pyqrcode
import png

# Data to encode
data = "https://encrypted-tbn1.gstatic.com/licensed-image?q=tbn:ANd9GcR6AK1OtJ2FXnHR6omrAJwzJUpDnlRN4K-L3NfsTyFLYE2M4MFb26tkd0lxmNiNcEtINJw_Bjvc3RrQysQ"

# Create QR code
qr = pyqrcode.create(data)

# Save as PNG
qr.png("qrcode.png", scale=8)

# Save as SVG
qr.svg("qrcode.svg", scale=8)
```

```python
# Import QRCode from pyqrcode
import pyqrcode
import png
from pyqrcode import QRCode


# String which represents the QR code
s = "https://encrypted-tbn1.gstatic.com/licensed-image?q=tbn:ANd9GcR6AK1OtJ2FXnHR6omrAJwzJUpDnlRN4K-L3NfsTyFLYE2M4MFb26tkd0lxmNiNcEtINJw_Bjvc3RrQysQ"

# Generate QR code
url = pyqrcode.create(s)

# Create and save the svg file naming "myqr.svg"
url.svg("myqr.svg", scale = 8)

# Create and save the png file naming "myqr.png"
url.png('myqr.png', scale = 6)
```
